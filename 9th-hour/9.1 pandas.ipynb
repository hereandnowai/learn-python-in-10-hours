# Using functions makes our code clean, but you can calculate the bmi without one
def calculate_bmi ():

    weights = df['Weight']

    heights = df['Height']

    bmi = []

    for w,h in zip(weights, heights):

        b = w/(h*h)

        bmi.append(b)

    return bmi
    
bmi = calculate_bmi()

df['BMI'] = bmi
